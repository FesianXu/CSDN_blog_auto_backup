```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
import time
import json
import urllib.request
import numpy as np
from bs4 import BeautifulSoup
import datetime
```

```python
csdn_login_url = 'https://passport.csdn.net/login?code=public'
opts = Options()
opts.add_argument("user-agent=Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.87 Safari/537.36")
```

```python
driver = webdriver.Chrome(chrome_options=opts)
```

```python
driver.get(csdn_login_url)
```

```python
def getHtml(url):
    html = urllib.request.urlopen(url).read()
    return html

def get_list(pages):
    alist = []
    for p in pages:
        fesian = 'https://blog.csdn.net/LoseInVain/article/list/{}'.format(p)
        pageSource = getHtml(fesian)
        soup = BeautifulSoup(pageSource)
        o = soup.find_all('div', {'data-articleid':True})
        
        for each in o:
            alist += [int(each['data-articleid'])]
    return alist

def get_markdown(articleids,mks=None):
    if mks == None:
        mks = {}
    for ind, each in enumerate(articleids):
        if each in mks:
            continue
        print(ind)
        name = 'https://mp.csdn.net/mdeditor/getArticle?id={}'.format(each)
        driver.get(name)
        pageSource = driver.page_source
        mks[each] = pageSource
        time.sleep(1)
    return mks

def get_time(articlesids):
    times = {}
    for each in articlesids:
        print(each)
        name = 'https://blog.csdn.net/LoseInVain/article/details/{}'.format(each)
        pageSource = getHtml(name)
        soup = BeautifulSoup(pageSource)
        update_time = soup.find_all('div', {'class': 'up-time'})[0].text[3:]
        times[each] = update_time
    return times

def split_time(datetime):
    date = datetime[:10].split('-')
    date = list(map(lambda x:int(x), date))
    time = datetime[11:].split(':')
    time = list(map(lambda x:int(x), time))
    return date, time

def write_to_files(save_path, times, file):
    s = BeautifulSoup(file)
    j = json.loads(s.text)
    strings = j['data']['markdowncontent'].split('\n')
    title = j['data']['title']
    thisid = int(j['data']['id'])
    date, time = split_time(times[thisid])
    title = title.replace('/','-')
    title = title.replace('|','')
    title = title.replace('<','《')
    title = title.replace('>','》')
    
    save_name = save_path+'title=({}),id=({}),uptime=({}-{}-{}-{}-{}-{}).md'.format(title, thisid, date[0], date[1], date[2],time[0], time[1], time[2])
#     print(save_name)
    f = open(save_name, 'w',encoding='utf-8')
    for each in strings:
        f.write(each)
        f.write('\n')
    f.close()
```

```python
alist = get_list((1,2,3,4)) # get the article id list
mks = get_markdown(alist) # get articles' markdown
times = get_time(articlesids=alist) # get the timestamp of each article
```

```python
for ind, (k,v) in enumerate(mks.items()):
    print(ind)
    try:
        write_to_files('./blog_mds/', times, mks[k])
    except Exception as e:
        print(e)
        print('the article is not written by markdown!')
        continue
```
